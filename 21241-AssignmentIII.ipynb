```python
from sklearn import preprocessing
label=preprocessing.LabelEncoder()
df_iris['Species']=label.fit_transform(df_iris['Species'])
df_iris
```

```python
df_iris.drop("Id",axis=1,inplace=True)
```

```python
df_iris
```

```python
from sklearn.preprocessing import StandardScaler
data=df_iris[['SepalLengthCm',"SepalWidthCm",'PetalLengthCm','PetalWidthCm']]
scaler=StandardScaler()
X = df_iris.copy()
X.iloc[:, [0,1,2,3]]= scaler.fit_transform(X.iloc[:,[0,1,2,3]])
```

```python
from sklearn.model_selection import train_test_split
x = X.iloc[:,:-1]
y = X.iloc[:,4]
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=39)
X_train.shape
```

#Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
model=LogisticRegression()
model.fit(X_train,y_train)
prediction=model.predict(X_test)
prediction
print('The accuracy of the Logistic Regression is',metrics.accuracy_score(prediction,y_test))
print('The precision of the Logistic Regression is',metrics.precision_score(prediction,y_test,average='micro'))
print('The recall of the Logistic Regression is',metrics.recall_score(prediction,y_test,average='micro'))
print('The f1 of the Logistic Regression is',metrics.f1_score(prediction,y_test,average='micro'))




```

```python
import seaborn as sn
```

```python
#Confusion Matrix to Calculate the accuracy of our Model
sn.set
sn.set(rc = {'figure.figsize':(10,5)})
confusion_matrix = pd.crosstab(y_test, prediction, rownames=['Actual'], colnames=['Predicted'])
sn.heatmap(confusion_matrix, annot=True)
print('Accuracy: ',metrics.accuracy_score(y_test, prediction))
plt.show()
```

```python
## Decision Boundary
from mlxtend.plotting import plot_decision_regions
value=0
width=1

plot_decision_regions(X_train.values, y_train.values, model, feature_index=[2,3], filler_feature_values={0: value, 1:value},
                      filler_feature_ranges={0: width, 1: width}, legend=3)
plt.ylabel('Petal Width (standardized)')
plt.xlabel('Petal Length (standardized)')
plt.title("Decision Boundary produced by Logistic Regression Model", size = 14)
```
